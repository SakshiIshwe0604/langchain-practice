```python
from langgraph.prebuilt import create_react_agent

agent_executor = create_react_agent(
    model=llm,
    tools=[search_tool]
)
```

```python
from langgraph.prebuilt import create_react_agent

agent_executor = create_react_agent(
    model=llm,
    tools=[search_tool]
)
```

```python
# Step 3: Create the ReAct agent with LLM and tools
# langgraph's create_react_agent handles the prompt internally
agent_executor = create_react_agent(
    model=llm,                              # the Groq LLM
    tools=[search_tool, get_weather_data]   # tools the agent can use
)
```

```python
# Step 4: No need for AgentExecutor вЂ” langgraph's create_react_agent
# already returns a ready-to-use executor directly
```

```python
# Step 5: Invoke the agent with a question
# langgraph uses "messages" format instead of "input"
response = agent_executor.invoke({
    "messages": [("user", "Find the capital of Madhya Pradesh, then find its current weather condition")]
})

# Print the final response
print(response["messages"][-1].content)
```
